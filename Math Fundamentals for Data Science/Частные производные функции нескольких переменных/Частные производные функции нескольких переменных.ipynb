## Задание 2. Минимизация среднеквадратической ошибки
Найдите наилучшую, с точки зрения среднеквадратичной ошибки, функцию, приближающую зависимость времени разгона автомобиля до 100 км/ч от его массы и мощности двигателя. 

```python
a2, a1, a0, = symbols('a2, a1, a0')
```

```python
MSE = 1/3 * (((a2 * 560 + a1 * 1945 + a0) - 4.3)**2 + \
             ((a2 * 340 + a1 * 1495 + a0) - 4.9)**2 + \
             ((a2 * 343 + a1 * 1570 + a0) - 5.2)**2)
MSE
```

```python
MSEa2 = diff(MSE, a2)
MSEa2
```

```python
MSEa1 = diff(MSE, a1)
MSEa1
```

```python
MSEa0 = diff(MSE, a0)
MSEa0
```

```python
nonlinsolve([MSEa2, MSEa1, MSEa0], [a2, a1, a0])
```
